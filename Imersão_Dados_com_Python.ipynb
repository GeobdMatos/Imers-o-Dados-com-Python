```python
renomear_colunas = {
    'work_year': 'ano_trabalho',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_empregado',
    'remote_ratio': 'taxa_remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'tamanho_empresa'
}

df = df.rename(columns=renomear_colunas)
print("Nomes das colunas após a tradução:")
print(df.columns)
```

```python
renomear_colunas = {
    'ano_trabalho': 'ano',
    'nivel_experiencia': 'senioridade',
    'tipo_emprego': 'contrato',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencia_funcionario': 'residencia',
    'taxa_remoto': 'remoto',
    'localizacao_empresa': 'empresa',
    'tamanho_empresa': 'tamanho_empresa'
}

df = df.rename(columns=renomear_colunas)

df.head()
```

```python
df['senioridade'].value_counts()
```

```python
df['contrato'].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
senioridade = {
    'SE': 'senior',
    'MI': 'pleno',
    'EN': 'junior',
    'EX': 'executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'integral',
    'PT': 'parcial',
    'CT': 'contrato',
    'FL': 'freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
tamanho_empresa = {
    'L': 'grande',
    'S': 'pequena',
    'M':	'media'

}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
mapa_trabalho = {
    0: 'presencial',
    100: 'remoto',
    50: 'hibrido'
}

df['remoto'] = df['remoto'].replace(mapa_trabalho)
df['remoto'].value_counts()
```

```python
df.describe(include='object')
```

```python
import numpy as np

df_salarios = pd.DataFrame({
  'nome': ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
  'salario': [4000, np.nan, 5000, np.nan, 100000],
})

df_salarios['salario_media'] = df_salarios["salario"].fillna(df_salarios["salario"].mean().round(2))
df_salarios['salario_mediana'] = df_salarios["salario"].fillna(df_salarios["salario"].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia' : ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura' : [30, np.nan, np.nan, 28, 27]
})
df_temperaturas['Preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'dia' : ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'temperatura' : [30, np.nan, np.nan, 28, 27]
})
df_temperaturas['Preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'cidade' : ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belem']

})
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```

# Aula 3 - Visualização de Dados


```python
df_limpo.head()
```

```python
df_limpo['senioridade'].value_counts().plot(kind='bar', title = 'Distribuição de senioridades')
```

```python
import seaborn as sns
```

```python
sns.barplot(data=df_limpo, x='senioridade', y='usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd')
plt.title('Salario medio por nivel de senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salario USD')
plt.show()
```

```python
df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=True).index
ordem
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd', order=ordem)
plt.title('Salario medio por nivel de senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salario USD')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df_limpo['usd'], bins = 50, kde=True)
plt.title('Distribuição dos salarios atuais')
plt.xlabel('Salario usd')
plt.ylabel('Frequencia')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_limpo['usd'])
plt.title('Boxplot salarios atuais')
plt.xlabel('Salario usd')
plt.show()
```

```python
ordem_senioridade = ['junior', 'pleno', 'senior', ' executivo']

plt.figure(figsize=(8, 4))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade)
plt.title('Boxplot salarios atuais')
plt.xlabel('Salario usd')
plt.show()
```

```python
ordem_senioridade = ['junior', 'pleno', 'senior', 'executivo']

plt.figure(figsize=(8,5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade, palette='Set2', hue='senioridade')
plt.title("Boxplot da distribuição por senioridade")
plt.xlabel("Salário em USD")
plt.show()
```

```python
import plotly.express as px
```

```python
senioridade_media_salario = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(senioridade_media_salario,
             x='senioridade',
             y='usd',
             title='Média Salarial por Senioridade',
             labels={'senioridade': 'Nível de Senioridade', 'usd': 'Média Salarial Anual (USD)'})

fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos tipos de trabalho'

          )

fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos tipos de trabalho',
             hole=0.5
          )
fig.update_traces(textinfo='percent+label')
fig.show()
```
